```python
import matplotlib.pyplot as plt
```

```python
file_name = "output.txt"
with open(file_name, "r", encoding="utf-8") as file:
    coords_x = []
    coords_y = []
    for i in file.read().splitlines():
        x, y = map(float, i.split())
        coords_x.append(x)
        coords_y.append(y)
```

```python
%matplotlib inline
plt.scatter(coords_x, coords_y, edgecolors="black")
plt.grid(True)
plt.show()
```
